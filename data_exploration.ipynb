```python
# import 200_baseline.csv into a pd df
df_baseline = pd.read_csv('200_baseline.csv', header=None)
df_baseline.head()
```

```python
# change column names to "timestamp", "second", "x", "y", "z"
df_baseline.columns = ['timestamp', 'second', 'x', 'y', 'z']
df_baseline.head()
```

```python
# print the number of rows in the dataframe
print(len(df_baseline))
```

```python
# extract all the rows for the column x in a list
x_baseline = df_baseline['x'].tolist()

# extract all the rows for the column y in a list
y_baseline = df_baseline['y'].tolist()

# extract all the rows for the column z in a list
z_baseline = df_baseline['z'].tolist()
```

```python
# plot the variation of x with time
plt.plot(x_baseline)
plt.show()
```

```python
# plot the variation of y with time
plt.plot(y_baseline)
plt.show()
```

```python
# plot the variation of z with time
plt.plot(z_baseline)
plt.show()
```

```python
# make a mel-spectrogram of z_experiment from 1650 to 3250
import librosa
import librosa.display
import numpy as np

# make a mel-spectrogram of z_experiment from 1650 to 3250
y, sr = librosa.load('Experiment', offset=0.165, duration=1.6)
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
log_S = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel')
plt.show()
```
